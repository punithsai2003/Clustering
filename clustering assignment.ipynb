```python
import pandas as pd
```

```python
data=pd.read_csv("EastWestAirlines.csv")
```

```python
data
```

```python
data.isnull().sum()
```

```python
data1=data.drop("ID#",axis=1)
```

```python
from sklearn.cluster import KMeans
```

# HIERARCHIAL CLUSTERING

```python
def norm_func(i):
    x=(i-i.min())/(i.max()-i.min())
    print(x)
    return(x)
```

```python
df_norm=norm_func(data)
```

```python
df_norm
```

```python
import scipy.cluster.hierarchy as sch 
from sklearn.cluster import AgglomerativeClustering

dendogram=sch.dendrogram(sch.linkage(df_norm, method='complete'))
```

```python
hc = AgglomerativeClustering(n_clusters=5, affinity = 'euclidean', linkage = 'complete',distance_threshold=None)
```

```python
hc.fit(df_norm)
```

```python
y_hc=hc.fit_predict(df_norm)
```

```python
y_hc
```

```python
data['h_clusterid']=y_hc
```

```python
data
```

# DB SCAN

```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
array=data.values
array
```

```python
stscaler= StandardScaler()
x=stscaler.fit_transform(array)
```

```python
from sklearn.neighbors import NearestNeighbors
import numpy as np
import matplotlib.pyplot as plt
nearest_neighbors = NearestNeighbors(n_neighbors=20)
neighbors = nearest_neighbors.fit(x)

distances, indices = neighbors.kneighbors(x)
distances=np.sort(distances[:,10],axis=0)
fig=plt.figure(figsize=(5, 5))
plt.plot(distances)
plt.xlabel("points")
plt.ylabel("Distance")
plt.grid()
```

```python
dbscan = DBSCAN(eps=1.30, min_samples=5)
clusters = dbscan.fit_predict(x)
```

```python
#Noisy samples are given the label -1.
dbscan.labels_
```

```python
data['cluster']=dbscan.labels_
data.head()
```
